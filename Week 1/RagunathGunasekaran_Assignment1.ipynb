## Loading Libraries

```python
import numpy as np
import pandas as pd

from numpy import array
from numpy import random
```

###  smaller array

```python
SmallerArray = random.choice(random.randint(1,500), size=(3, 2))
print(SmallerArray)
```

```python
print("Sum of all elements:")
print(np.sum(SmallerArray))
print("Sum of each column:")
print(np.sum(SmallerArray, axis=0))
print("Sum of each row:")
print(np.sum(SmallerArray, axis=1))
```

```python
print("Min of all elements:")
print(np.min(SmallerArray))
print("Max of all elements:")
print(np.max(SmallerArray))
print("average of all elements:")
print(np.average(SmallerArray))
print("mean of all elements:")
print(np.mean(SmallerArray))
print("std of all elements:")
print(np.std(SmallerArray))
```

```python
SmallerArray_df = pd.DataFrame(data = SmallerArray) 
SmallerArray_df.columns=["Col "+ str(i+1) for i in range(0, 2)]
SmallerArray_df.index =["index "+ str(i+1) for i in range(0, 3)]
SmallerArray_df
```

## 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
#  choice()  returns a randomly selected element from the specified sequence
#  randint() returns an integer number selected element from the specified range 1 to 500

ArrayAll = random.choice(random.randint(2,500), size=(10, 50))
print(ArrayAll)
```

## 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
print("Sum of all elements:")
print(np.sum(ArrayAll))
print("Sum of each column:")
print(np.sum(ArrayAll, axis=0))
print("Sum of each row:")
print(np.sum(ArrayAll, axis=1))
```

## 3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
print("Min of all elements:")
print(np.min(ArrayAll))
print("Max of all elements:")
print(np.max(ArrayAll))
print("average of all elements:")
print(np.average(ArrayAll))
print("mean of all elements:")
print(np.mean(ArrayAll))
print("std of all elements:")
print(np.std(ArrayAll))
```
