```python
# color for each label
colors = ['r', 'y', 'g', 'b']
# plotting the values into Bar Chart 
Income_2018 =Income_df["2018_Income"]
Income_2019 =Income_df["2019_Income"]
Region= Income_df["Region"]
```

## Bar Chart

```python
plt.bar(Region, Income_2018,align='center', alpha=0.5, color =colors)
# naming the x axis
plt.xlabel('US Region')
# naming the Y axis
plt.ylabel('2018 Income in Thousands (Dollars)')
# naming the Title
plt.title('2018 Income at Each Region in USA - BAR Chart')
# setting range
#Show the visual
plt.show()
```

## Pie Chart

```python
# plotting the values into Pie Chart 
plt.pie(Income_2018, labels=Region, colors=colors,autopct='%1.1f%%', 
    shadow=True, startangle=90)
# naming the Title
plt.title('2018 Income in Thousands (Dollars) - PIE Chart')
plt.axis('equal')
# plotting legend
plt.legend()
# show the chart
plt.show()
```

## Line Plot

```python
plt.plot(Region, Income_2018, color='darkblue', marker='x', label="2018")
# naming the x axis
plt.xlabel('US Region')
# naming the Y axis
plt.ylabel('2018 Income in Thousands (Dollars)')
# naming the Title
plt.title('2018 Income at Each Region in USA - Line Chart')
plt.grid(True)
```

## Scatter Plot

```python
plt.scatter(Region, Income_2018, color='Red', marker='x', label="2018")
# naming the x axis
plt.xlabel('US Region')
# naming the Y axis
plt.ylabel('2018 Income in Thousands (Dollars)')
# naming the Title
plt.title('2018 Income at Each Region in USA - scatter Chart')
plt.grid(True)
```

Relationship - Scatter Plot, Bubble sort

Composition - Pie, Tree Map for Static and Stacked Bar charts, area chart for Changing data
Comparison of data - Bar, Line Chart


In our dataset contains data for both 2018 and 2019 for US Regions. Hence I prefer to show how the income data vary per each Region.


Bar Chart
It is most effective when we are trying to visualize categorical data that has few (probably < 10) categories. In our data set we have 4 US Regions.
In the barplot() function, x_data represents the US Regions on the x-axis and Income Data represents the bar height on the y-axis.

Hence I rate BAR Chart as first preference for this plot

Pie Chart
As we know, Pie chart used for when demonstrating the proportional composition of a particular variable over a static timeframe. The Good part with Pie Chart is we can clearly understand how as a whole each Region performs with percentage.
when I look at Pie chart, I can easly tell the South Region has more income almost 40%

Hence I rate BAR Chart as Second preference for this plot

Line Plots 
Line plots are best used when you can clearly see that one variable varies greatly with another (Trending). I rated the Line chart as third preference because each Region income is varying very less.
It will become diffuclt to understand very clearly how each Region varies with each other.

Hence I rate Line Chart as Third preference for this plot

Scatter Plots 
Scatter plots are great for showing the relationship between two variables since you can directly see the raw distribution of the data.
